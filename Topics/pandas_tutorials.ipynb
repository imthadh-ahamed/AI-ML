## Data Frames

A DataFrame is a **two-dimensional, size-mutable, and potentially heterogeneous tabular** data structure with labeled axes (rows and columns).<br>
It's similar to a table in a database or an Excel spreadsheet.<br><br>

## Key Features of DataFrames:
**Labeled Axes:** Both rows and columns have labels, which makes data manipulation and retrieval easy.<br>
**Heterogeneous Data:** Columns can contain different types of data (integers, floats, strings, etc.).<br>
**Size-Mutable:** We can add or delete columns and rows.

```python
data = {
    'Name' : ['Kamal', 'Vimal', 'Jhon', 'Stephen'],
    'Age' : [10, 25, 45, 12],
    'City' : ['NY', 'CLMB', 'TN', 'UK']
}

df = pd.DataFrame(data)
df
```
